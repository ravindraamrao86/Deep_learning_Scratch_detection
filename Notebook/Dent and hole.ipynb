```python
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator

def check_directory(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"The directory {path} does not exist.")
    if not os.listdir(path):
        raise ValueError(f"The directory {path} is empty.")

# Model Creation
def create_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(3, activation='softmax'))  # 3 classes: scratches, dents, holes

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Check if directories exist and are not empty
train_dir = r'C:\Users\Laptop On Rent 078\Deep_learning_Scratch_detection\Notebook\Datasets\Train'
val_dir = r'C:\Users\Laptop On Rent 078\Deep_learning_Scratch_detection\Notebook\Datasets\test'

check_directory(train_dir)
check_directory(val_dir)

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

try:
    training_set = train_datagen.flow_from_directory(
        train_dir,
        target_size=(128, 128),
        batch_size=32,
        class_mode='categorical'
    )
    test_set = test_datagen.flow_from_directory(
        val_dir,
        target_size=(128, 128),
        batch_size=32,
        class_mode='categorical'
    )
except Exception as e:
    print(f"Error loading data: {e}")
    raise

# Print the model summary to ensure the output layer is correct
model = create_model()
model.summary()

# Print the shapes of the data batches
for data_batch, labels_batch in training_set:
    print("Training data batch shape:", data_batch.shape)
    print("Training labels batch shape:", labels_batch.shape)
    break

for data_batch, labels_batch in test_set:
    print("Validation data batch shape:", data_batch.shape)
    print("Validation labels batch shape:", labels_batch.shape)
    break

# Train the model
history = model.fit(
    training_set,
    validation_data=test_set,
    epochs=5,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set),
    verbose=1
)
```
